```python
off_df = pd.concat([df2017, df2018, df2019])
off_df
```

```python
off_df['Tm'].replace('Minnesota Vikings','MIN',inplace=True)
off_df['Tm'].replace('Jacksonville Jaguars','JAX',inplace=True)
off_df['Tm'].replace('Los Angeles Chargers','LAC',inplace=True)
off_df['Tm'].replace('Philadelphia Eagles','PHI',inplace=True)
off_df['Tm'].replace('New England Patriots','NE',inplace=True)
off_df['Tm'].replace('Baltimore Ravens','BAL',inplace=True)
off_df['Tm'].replace('Pittsburgh Steelers','PIT',inplace=True)
off_df['Tm'].replace('Atlanta Falcons','ATL',inplace=True)
off_df['Tm'].replace('Chicago Bears','CHI',inplace=True)
off_df['Tm'].replace('New Orleans Saints','NO',inplace=True)
off_df['Tm'].replace('Carolina Panthers','CAR',inplace=True)
off_df['Tm'].replace('Los Angeles Rams','LA',inplace=True)
off_df['Tm'].replace('Dallas Cowboys','DAL',inplace=True)
off_df['Tm'].replace('Seattle Seahawks','SEA',inplace=True)
off_df['Tm'].replace('Cincinnati Bengals','CIN',inplace=True)
off_df['Tm'].replace('Tennessee Titans','TEN',inplace=True)
off_df['Tm'].replace('Buffalo Bills','BUF',inplace=True)
off_df['Tm'].replace('Arizona Cardinals','ARI',inplace=True)
off_df['Tm'].replace('Oakland Raiders','LV',inplace=True)
off_df['Tm'].replace('Detroit Lions','DET',inplace=True)
off_df['Tm'].replace('New York Jets','NYJ', inplace=True)
off_df['Tm'].replace('Denver Broncos','DEN',inplace=True)
off_df['Tm'].replace('Tampa Bay Buccaneers','TB',inplace=True)
off_df['Tm'].replace('San Francisco 49ers','SF',inplace=True)
off_df['Tm'].replace('Green Bay Packers','GB',inplace=True)
off_df['Tm'].replace('New York Giants','NYG',inplace=True)
off_df['Tm'].replace('Washington Redskins','WAS',inplace=True)
off_df['Tm'].replace('Miami Dolphins','MIA',inplace=True)
off_df['Tm'].replace('Indianapolis Colts','IND',inplace=True)
off_df['Tm'].replace('Cleveland Browns','CLE',inplace=True)
off_df['Tm'].replace('Houston Texans','HOU',inplace=True)
off_df
```

```python
off_df.info()
```

```python
off_df['TD_PA'] = off_df['TD']/off_df['Att']
off_df['INT_PA'] = off_df['Int']/off_df['Att']
off_df['1D_PA'] = off_df['1stD.1']/off_df['Att']
off_df['TD_RA'] = off_df['TD.1']/off_df['Att.1']
off_df['1D_RA'] = off_df['1stD.2']/off_df['Att.1']
off_df['Fum_Play'] = off_df['FL']/off_df['Ply']
off_df['Comp_Pct'] = off_df['Cmp']/off_df['Att']
```

```python
off_df_final = off_df[['Tm', 'Year', 'PF', 'Fum_Play', 'Att', 'Comp_Pct', 'TD_PA', 'INT_PA', 'NY/A', '1D_PA', 'Att.1', 'TD_RA', 'Y/A', '1D_RA']]
off_df_final
```

```python
off_df_final = off_df_final.rename(columns={'Att':'Att_Pass', 'Att.1':'Att_Rush'})
```

```python
off_df_final = off_df_final.add_prefix('Off_')
off_df_final
```
