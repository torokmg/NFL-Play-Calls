```python
import pandas as pd
```

```python
YEARS = [2019,2018,2017]
```

```python
data = pd.DataFrame()

for i in YEARS:  
    i_data = pd.read_csv('https://github.com/guga31bb/nflfastR-data/blob/master/data/' \
                         'play_by_play_' + str(i) + '.csv.gz?raw=True',
                         compression='gzip', low_memory=False)

    data = data.append(i_data, sort=True)
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
cols = data.columns.tolist()
```

```python
cols
```

```python
df = data[['game_date', 'posteam', 'defteam', 'posteam_type', 'yardline_100','quarter_seconds_remaining','half_seconds_remaining','game_seconds_remaining','qtr','down','goal_to_go','ydstogo','play_type','score_differential','shotgun','no_huddle']]
df = df.dropna()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df['posteam_type']=df['posteam_type'].map({'home': '1', 'away': '0'}).astype(int)
```

```python
df
```

```python
df.info()
```

```python
play_df = df[df['play_type'].isin(['run', 'pass'])]
```

```python
pd.to_datetime(play_df.game_date, infer_datetime_format=True)
```

```python
play_df['game_date'] = pd.DatetimeIndex(play_df['game_date']).year
play_df
```

```python
play_df = play_df.rename(columns={'game_date':'Year'})
```

```python
play_df['Year']=play_df['Year'].astype(int)
```

```python
play_df['Year']=play_df['Year'].map({2020: 2019, 2019:2019, 2018:2018, 2017:2017})
```
