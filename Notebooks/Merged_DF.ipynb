# Merging Into a Final DF

Now that we have created our 3 seperate dataframes, we will import them and merge them into a single dataframe to model off of.

```python
# Import Library
import pandas as pd
```

```python
# Import previously created .csv files
def_df = pd.read_csv('Data/def_df.csv')
play_df = pd.read_csv('Data/play_df.csv')
off_df = pd.read_csv('Data/off_df.csv')
```

We will merge our dataframes in two steps using left joins merged on both the year and team names:

* First we will merge the play_df with our off_df 
* Second we will merge the above merged df with our def_df 

```python
# Merge all dfs
final_df = pd.merge(play_df, off_df,  how='left', left_on=['Year','posteam'], right_on = ['Off_Year','Off_Tm'])
final_df = pd.merge(final_df, def_df,  how='left', left_on=['Year','defteam'], right_on = ['Def_Year','Def_Tm'])
```

```python
final_df
```

```python
final_df.info()
```

Now we can drop the team name and year columns from the dataframe as they were already displayed from our original play_df.

```python
# Drop redundant columns
final_df = final_df.drop(columns={'Off_Tm', 'Off_Year', 'Def_Tm', 'Def_Year'})
final_df.info()
```
